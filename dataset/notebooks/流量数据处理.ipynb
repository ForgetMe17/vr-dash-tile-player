```python
import numpy as np
import json
import matplotlib
import matplotlib.pyplot as plt
```

```python
def str2array(str):
    m_data = str.split(',')
    return [float(n) for n in m_data]
```

```python
with open("backend-data.txt", 'r') as f:
    data = json.load(f)
throughput = data['throughput']
timeline = data['timeline']
plt.plot(timeline, throughput)
plt.axis([0,30,0,2e9])
```

```python
with open("viewerData.json", 'r') as f:
    fovdata = json.load(f)
user4fov = np.array(fovdata['user_4'])
```

```python
combine_num = 300
cut_len = int((user4fov.shape[0])/combine_num) * combine_num
x = user4fov[1:cut_len+1, 0].reshape(-1, combine_num)
y = abs(np.diff(user4fov[:cut_len+1, 1])).reshape(-1, combine_num)
x = x[:, 0]
y = np.mean(y, axis=1)
plt.figure()
plt.plot(timeline, throughput)
plt.axis([0,120,0,2e9])
plt.figure()
plt.plot(x,y)
plt.axis([0,120, 0, 0.016])
```
